```python
class User:
    def __init__(self, first_name, last_name, birth_year):
        self.first_name = first_name
        self.last_name = last_name
        self.birth_year = birth_year
    
    def information(self):
        print(f"Name: {self.first_name} {self.last_name}, Birth year: {self.birth_year}")


n = int(input())
a = []
for i in range(n):
    first_name, last_name, birth_year = input().split()
    u = User(first_name, last_name, birth_year)
    a.append(u)

for u in a:
    u.information()

import pydoc
print(pydoc.render_doc(a[0], renderer=pydoc.plaintext), dir(a[0]))
```

```python
class User:
    total_users = 0
    def __init__(self, first_name, last_name, birth_year):
        self.first_name = first_name
        self.last_name = last_name
        self.birth_year = birth_year

        User.total_users += 1
    
    def information(self):
        return f"Name: {self.first_name} {self.last_name}, Birth year: {self.birth_year}"


n = int(input())
users = dict()
for i in range(n):
    first_name, last_name, birth_year = input().split()
    u = User(first_name, last_name, birth_year)
    users.update({User.total_users: u})

for key, val in users.items():
    print(f"ID: {key}, {val.information()}")

print(f"Total users: {User.total_users}")

import pydoc
print(pydoc.render_doc(u, renderer=pydoc.plaintext), dir(u))
```
