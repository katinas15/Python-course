Fibonačio skaičiai gali būti apskaičiuojami naudojant šią formulę:

Naudodami numpy veiksmus su matricomis raskite n-tąjį Fibonačio skaičių, jeigu F₀ = 0, F₁ = 1.

```python
import numpy as np
np.set_printoptions(suppress=True)
n = int(input())
f0 = 0
f1 = 1
v1 = np.vstack(np.array([[1, 1], [1, 0]], dtype=np.uint64))
daug = np.array([[f1], [f0]], dtype=np.uint64)
powered = np.power(v1, n-1)
a = np.linalg.matrix_power(v1, n-1) * daug
print(a[0,0])
```
